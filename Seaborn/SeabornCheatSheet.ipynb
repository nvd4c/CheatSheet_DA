# Seaborn

#Thư viện trực quan hóa Python Seaborn dựa trên matplotlib và cung cấp giao diện cấp cao để vẽ đồ họa thống kê hấp dẫn. 

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

#The basic steps to creating plots with Seaborn are:
 1. Chuẩn bị một số dữ liệu
 2. Kiểm soát tính thẩm mỹ của con số
 3. Plot với seaborn
 4. Customize plot

```python
tips = sns.load_dataset("tips")                                 # step 1 Chuẩn bị dữ liệu
sns.set_style("whitegrid")                                      # step 2 Kiểm soát tính thẩm mỹ của con số
g = sns.lmplot(x="tip", y="total_bill", data=tips, aspect=2)    # step 3 Plot some Seaborn #tip: tiền
g = (g.set_axis_labels("Tip","Total bill(USD)").set(xlim=(0,10),ylim=(0,100)))
plt.title("title")                                              # step 4 Customize plot
plt.show(g)
```
