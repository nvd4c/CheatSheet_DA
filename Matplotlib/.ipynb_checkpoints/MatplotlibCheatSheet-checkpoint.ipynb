# Matplotlib
là một thư viện vẽ đồ thị 2D của Python tạo ra số liệu chất lượng xuất bản ở nhiều định dạng bản cứng và các môi trường tương tác trên nền tảng

# 1. Prepare The Data

# 1.1 1D Data


```python
 import numpy as np
```

```python
x = np.linspace(0, 10, 100) #trả về các số cách đều nhau trong 1 khoảng
y = np.cos(x) 
z = np.sin(x)
```

# 1.2 2D Data

```python
data = 2 * np.random.random((10, 10))
data2 = 3 * np.random.random((10, 10))
Y, X = np.mgrid[-3:3:100j, -3:3:100j]
U = -1 - X**2 + Y
V = 1 + X - Y**2
from matplotlib.cbook import get_sample_data
img = np.load(get_sample_data('axes_grid/bivariate_normal.npy'))
```

# 2. Create plot

```python
import matplotlib.pyplot as plt
```

# 2.1 figure

```python
fig = plt.figure()
fig2 = plt.figure(figsize=plt.figaspect(2.0))
```

# 2.2 Trục
Tất cả các plot được thực hiện đối với Axes. Trong hầu hết các trường hợp, một
subplot sẽ phù hợp với nhu cầu của bạn. Một subplot là một trục trên hệ thống lưới.

```python
fig.add_axes()
ax1 = fig.add_subplot(221) # row-col-num
ax3 = fig.add_subplot(212) 
fig3, axes = plt.subplots(nrows=2,ncols=2)
fig4, axes2 = plt.subplots(ncols=3)
```

# 3. Quy trình lập plot

# 3.1 1D Data

```python
fig, ax = plt.subplots()
lines = ax.plot(x,y)                   #Vẽ các điểm bằng các đường hoặc điểm đánh dấu kết nối chúng
ax.scatter(x,y)                        #Vẽ các điểm không được kết nối, chia tỷ lệ hoặc tô màu
axes[0,0].bar([1,2,3],[3,4,5])         #Vẽ các hình chữ nhật theo chiều dọc (chiều rộng không đổi) 
axes[1,0].barh([0.5,1,2.5],[0,1,2])    #Vẽ các hình chữ nhật nằm ngang (chiều cao không đổi)
axes[1,1].axhline(0.45)                #Vẽ một đường ngang qua các trục 
axes[0,1].axvline(0.65)                #Vẽ một đường thẳng đứng qua các trục
ax.fill(x,y,color='blue')              #Vẽ đa giác đã lấp đầy 
ax.fill_between(x,y,color='yellow')    #Điền vào giữa các giá trị y và 0
```

# 3.1.1 Vector Fields

```python
axes[0,1].arrow(0,0,0.5,0.5)       #Thêm mũi tên vào các trục
#axes[1,1].quiver(y,z) Plot a       #Trường mũi tên 2D
axes[0,1].streamplot(X,Y,U,V)      #Vẽ một trường mũi tên 2D
```

# 3.1.2 Phân phối dữ liệu

```python
ax1.hist(y)                        #Vẽ biểu đồ
ax3.boxplot(y)                     #Make a box and âm mưu râu ria
ax3.violinplot(z)                  #Make a violin plot
```

# 3. 2 2D Data or Images

```python
fig, ax = plt.subplots()
im = ax.imshow(img, cmap='gist_earth', interpolation='nearest', vmin=-2, vmax=2)      #Mảng RGB hoặc được ánh xạ màu
```

```python
axes2[0].pcolor(data2)                 #Biểu đồ giả của mảng 2D
axes2[0].pcolormesh(data)              #Biểu đồ giả của mảng 2D
CS = plt.contour(Y,X,U)                #Vẽ đường bao
#axes2[2].contourf(data1)               #Vẽ đường bao đã điền
axes2[2]= ax.clabel(CS)                #Label a contour plot
```

# Phân tích plot và quy trình làm việc 

# The basic steps to creating plots with matplotlib are: 
1 Prepare data 
2 Create plot 
3 Plot 
4 Customize plot 
5 Save plot 6Show plot

```python
import matplotlib.pyplot as plt
x = [1,2,3,4]                                               #step1: prepare data
y = [10,20,25,30]
fig = plt.figure()                                          #step2: create plot
ax = fig.add_subplot(111)                                   #step 3: plot
ax.plot(x, y, color='lightblue', linewidth=3)               #step 3,4: plot and customize plot
ax.scatter([2,4,6], [5,15,25], color='darkgreen', marker='^')
ax.set_xlim(1, 6.5)
plt.savefig('foo.png')
plt.show()                                                  #step 6: show plot 
```

# 4.6 Limits, Legends & Layouts

```python
#Limits & Autoscaling
ax.margins(x=0.0,y=0.1)                                                     #Thêm phần padding vào 1 plot
ax.axis('equal')                                                            #Set the aspect ratio of the plot to 1
ax.set(xlim=[0,10.5],ylim=[-1.5,1.5])                                       #Đặt giới hạn cho trục x và trục y
ax.set_xlim(0,10.5)                                                         #Đặt giới hạn cho trục x

#Legends
ax.set(title='An Example Axes', ylabel='Y-Axis', xlabel='X-Axis')           #Đặt tiêu đề và các nhãn trục x và trục y
ax.legend(loc='best')                                                       #Không chồng chéo các plot

#Ticks
ax.xaxis.set(ticks=range(1,5), ticklabels=[3,100,-12,"foo"])                #Đặt x-tick theo cách thủ công
ax.tick_params(axis='y', direction='inout', length=10)                      #Làm cho y-tick dài hơn và ra vào

#Subplot Spacing
fig3.subplots_adjust(wspace=0.5, hspace=0.3, left=0.125, right=0.9, top=0.9, bottom=0.1) #Điều chỉnh khoảng cách giữa các plot phụ
fig.tight_layout()                                                                       #Điều chỉnh (các) ô con vào vùng hình

#Axis Spines
ax1.spines['top'].set_visible(False)                            #trục trên cùng cho một plot ẩn 
ax1.spines['bottom'].set_position(('outward',10))               #Di chuyển đường trục dưới cùng ra ngoài
```
